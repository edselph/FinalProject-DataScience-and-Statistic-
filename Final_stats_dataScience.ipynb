```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd 
from scipy import stats
```

```python
# Reading the CSV file
df = pd.read_csv('Indonesia_Covid_cases.csv')
```

```python
# Making the Days as the X-axis and the Total Covid-19 case number as the Y-axis
x = df['total_cases']
y = df['total_deaths']
```

```python
# Ploting the scatter graph
plt.figure(figsize=(17,7))
new_cfm = plt.scatter(x,y, s = 10)
plt.legend(['death count'])
plt.title('Mortality Over Covid-19 Cases')
plt.ylabel("Mortality count")
plt.xlabel("Total cases")
plt.title('Mortality Over Covid-19 Cases in Indonesia')
plt.grid(True)
plt.show()
```

```python
# Doing linear regression to get the best fit line to make our prediction
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.figure(figsize=(17,7))
plt.grid(True)
plt.scatter(x, y, s = 10)
plt.plot(x, mymodel, color = "r")
plt.ylabel("Mortality count")
plt.xlabel("Total cases")
plt.legend(['best fit','death count'])
plt.title('Mortality Over Covid-19 Cases in Indonesia')
plt.show()
```

```python
# printing out the values of the variables
print('slope =', slope,
      'intercept =', intercept,
      'r =', r,
      'p =', p,
      'std_err =', std_err)
```

```python
# allowing the user to enter mortality number to see the total number of the covid case at that day
print('Enter Covid-19 case number: ')
slope, intercept, r, p, std_err = stats.linregress(x, y)
i = int(input())
def myfunc(u):
  return slope * u + intercept
case_num = myfunc(i)
print('Total number of Covid-19 cases in Indonesia is', case_num )
```

```python
# allowing the user to enter the covid case numeber to see the number of death
print('Enter death count: ')
slope, intercept, r, p, std_err = stats.linregress(x, y)
i = int(input())
def myfunc(u):
  return (u - intercept)/ slope
death_num = myfunc(i)
print('number of death', death_num)
```

```python
#R_Square and Adjusted R Square
X_addC = sm.add_constant(x)
result = sm.OLS(y, X_addC).fit()
print(result.rsquared, result.rsquared_adj)
```
